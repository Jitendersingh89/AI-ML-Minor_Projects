## Pre-processing

### 1) converting to lowercase

```python
df["review"] = df["review"].str.lower()
```

### 2) remove the url

```python
# using the library of a python to do so re(regular expression)
#library - re
# sub function - used for substitute
import re

#e.g->
#sample_text = "abc is the word" # abc => xyz
#new_text = re.sub("abc" , "xyz" , sample_text)
```

```python
def remove_url(text):
    text = re.sub( r"http\S+" , "" , text)
    # (pattern , replace , strin)
    return text
df["review"] = df["review"].apply(remove_url)   
# \S vs \s 
# \s - included white space 
#\S - not included the white space 
```

### 3). removing the punctuations

```python
def remove_punctuations(text):
    text = re.sub( r"^[A-Za-z0-9\s]" , "" , text) 
    return text 

df["review"] = df["review"].apply(remove_punctuations)
```

### 4)removing HTML tags 

```python
### 4)removing HTML tags 
def remove_html(text):
    text = re.sub( r"<.*?>" , "" , text) 
    return text 

df["review"] = df["review"].apply(remove_html)
```

### 5) Removing the Stopwords

```python
# we gonna use the nlkt library
# firstly we have to toeknized the whole sentence 

import nltk 
nltk.download("punkt")# sentence tokenizer 
nltk.download("punkt_tab")
nltk.download("stopwords")
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords 
```

```python
#sample_text = "I like coding in python !"
#tokens = word_tokenize(sample_text)
```

```python
# function to remove the stop words

def remove_stopwords(text):
    tokens = word_tokenize(text) # tokeized the words 
    stop_words = stopwords.words("english") #stopwords of all the english 


    for word in tokens:
        if word in stop_words:
            text = text.replace(word , "")

    return text 

df["review"] = df["review"].apply(remove_stopwords)
```

## 6. stemming 

```python
# running - run
# played - play
# this by using porterstemming algorithm

from nltk.stem import PorterStemmer
```

```python
def stemming(text):
    ps = PorterStemmer()
    stemmed_words = []

    tokens = word_tokenize(text)
    for token in tokens:
        stemmed_token = ps.stem(token)
        stemmed_words.append(stemmed_token)

    return " ".join(stemmed_words)

df["review"]= df["review"].apply(stemming)
```
